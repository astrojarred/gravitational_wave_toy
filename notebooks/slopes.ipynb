# Investingating the slopes of the spectra

```python
from pathlib import Path
from gravitational_wave_toy import sensitivity, observe
import random
from matplotlib import pyplot as plt
import astropy.units as u
```

```python
catalog = Path("/Users/jarred/Downloads/TeV Catalog O5")
files = list(catalog.glob("*.fits"))
```

```python
for f in random.choices(files, k=1000):
    g = observe.GRB(f)
    p = g.show_spectral_evolution(return_plot=True)
    
plt.show()
```

```python
indices = []

for f in random.choices(files, k=1000):
    g = observe.GRB(f)
    indices.append(g.get_spectral_index(100 * u.s))
```

```python
plt.hist(indices, bins=20)
```

```python
indices = []

for f in random.choices(files, k=1000):
    g = observe.GRB(f)
    indices.append(g.get_spectral_index(10 * u.s))
```

```python
plt.hist(indices, bins=20)
```

```python
indices = []

for f in random.choices(files, k=1000):
    g = observe.GRB(f)
    i = g.get_spectral_index(1000 * u.s)
    indices.append(i)
    if i < -3.2:
        print(g.filepath)
    
plt.hist(indices, bins=20)
```
